###  Feature Engineering

```python
# Q11: Create a feature "Boundary Percentage

n['Boundary_Runs'] = n['Fours'] * 4 + n['Sixes'] * 6

n['Boundary_Percentage'] = (n['Boundary_Runs'] / n['Runs_Scored']) * 100
n['Boundary_Percentage'] = n['Boundary_Percentage'].round(2)

n[['Player_Name', 'Year', 'Runs_Scored', 'Boundary_Runs', 'Boundary_Percentage']].head()
```

```python
# Q12: Create a Batting Impact Score
n['Batting_Impact_Score'] = (n['Batting_Average'] * 0.4) + (n['Batting_Strike_Rate'] * 0.6)
n['Batting_Impact_Score'] = n['Batting_Impact_Score'].round(2)

# Top 5 by impact
n[['Player_Name', 'Year', 'Batting_Impact_Score']].sort_values(by='Batting_Impact_Score', ascending=False).head()
```

```python
# Q13: All-Rounder Performance Metric

n['Bowling_Impact_Score'] = n['Wickets_Taken'] * (1 / (n['Economy_Rate'] + 1))  # add 1 to avoid divide by zero
n['Bowling_Impact_Score'] = n['Bowling_Impact_Score'].round(2)

# Combine both for All-Rounder Score
n['All_Rounder_Score'] = n['Batting_Impact_Score'].fillna(0) + n['Bowling_Impact_Score'].fillna(0)

# Top 5 all-rounders
n[['Player_Name', 'Year', 'All_Rounder_Score']].sort_values(by='All_Rounder_Score', ascending=False).head()
```

```python
# Q14: Categorize players as Aggressive, Balanced, or Defensive

def categorize_player(sr):
    if pd.isna(sr):
        return 'Unknown'
    elif sr >= 140:
        return 'Aggressive'
    elif sr >= 110:
        return 'Balanced'
    else:
        return 'Defensive'

n['Player_Type'] = n['Batting_Strike_Rate'].apply(categorize_player)

n['Player_Type'].value_counts()
```

```python
# Q15: Compute year-over-year growth in key performance metrics

n.sort_values(by=['Player_Name', 'Year'], inplace=True)

n['YOY_Runs_Growth'] = n.groupby('Player_Name')['Runs_Scored'].pct_change().round(2)
n['YOY_Average_Growth'] = n.groupby('Player_Name')['Batting_Average'].pct_change().round(2)
n['YOY_SR_Growth'] = n.groupby('Player_Name')['Batting_Strike_Rate'].pct_change().round(2)

n[n['Player_Name'] == 'Virat Kohli'][['Year', 'Runs_Scored', 'YOY_Runs_Growth']]
```

```python
# Compute the year-over-year growth in performance metrics for each player.
n = n.sort_values(by=["Player_Name", "Year"])

n["YoY_Runs_Growth"] = n.groupby("Player_Name")["Runs_Scored"].pct_change() * 100

n["YoY_StrikeRate_Growth"] = n.groupby("Player_Name")["Batting_Strike_Rate"].pct_change() * 100
n["YoY_Wickets_Growth"] = n.groupby("Player_Name")["Wickets_Taken"].pct_change() * 100
n.head()
```

### Python-Based Queries

```python
# Q21: Get top 5 players with the highest Batting Average

top_avg = n[['Player_Name', 'Batting_Average']].dropna()
top_avg = top_avg.groupby('Player_Name').max().sort_values(by='Batting_Average', ascending=False).head(5)
print("Top 5 Players by Batting Average:")
print(top_avg)
```

```python
# Q22: Find the player who has played the most matches

n['Total_Matches'] = n[['Matches_Batted', 'Matches_Bowled']].fillna(0).sum(axis=1)
most_matches = n.groupby('Player_Name')['Total_Matches'].sum().sort_values(ascending=False).head(1)
print("Player with Most Matches Played:")
print(most_matches)
```

```python
# Q23: Retrieve all players who have scored more than 500 runs in a season

high_scorers = n[n['Runs_Scored'] > 500][['Player_Name', 'Year', 'Runs_Scored']]
print("Players who scored more than 500 runs in a season:")
print(high_scorers)
```

```python
# Q24: Identify players who have taken at least one five-wicket haul

five_hauls = n[n['Five_Wicket_Hauls'] >= 1][['Player_Name', 'Year', 'Five_Wicket_Hauls']]
print("Players with 5-wicket hauls:")
print(five_hauls)
```
